```python
import matplotlib.pyplot as plt
import pandas as pd

file_path = "C:/Users/user/Downloads/grandprix_winners_cleaned.csv"
data = pd.read_csv(file_path)

data['Date'] = pd.to_datetime(data['Date'], format='%d %b %Y')
def time_to_seconds(time_str):
    parts = time_str.split(':')
    if len(parts) == 3:
        return int(parts[0]) * 3600 + int(parts[1]) * 60 + float(parts[2])
    elif len(parts) == 2:
        return int(parts[0]) * 60 + float(parts[1])
    return float(time_str)

data['Time_in_seconds'] = data['Time'].apply(time_to_seconds)
plt.figure(figsize=(10, 6))
plt.plot(data['Date'], data['Time_in_seconds'], marker='o', linestyle='-', color='coral')
plt.title('Race Duration Over Time', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Race Duration (Seconds)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = "C:/Users/user/Downloads/fastest_laps_cleaned.csv"
data = pd.read_csv(file_path)
sns.set_theme(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.histplot(data['time_seconds'], bins=20, kde=True, color='blue')
plt.title('Distribution of Lap Times (in Seconds)', fontsize=16)
plt.xlabel('Lap Time (seconds)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.show()
```
