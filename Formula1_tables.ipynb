```python
columns_practice = ["Grand Prix","Pos","No","Driver","Car","Time","Gap","Laps"
]
```

```python
def scrape_practice_data(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    # Extract Grand Prix name from the URL or page header
    grand_prix = url.split('/')[-3].replace('-', ' ').title()

    # Find the table in the page
    table = soup.find('table', class_='f1-table f1-table-with-data w-full')
    rows = table.find_all('tr')[1:]  # Skip the header row

    # List to hold row data
    data_practice = []

    # Extract data from each row
    for row in rows:
        cols = row.find_all('td')
        cols = [col.text.strip() for col in cols]
        data_practice.append([grand_prix] + cols)

    return data_practice
```

```python
all_practice_data = []
```

```python
for url in urls_practice:
    try:
        practice_data = scrape_practice_data(url)
        all_practice_data.extend(practice_data)
        print(f"Scraped data for {url}")
    except Exception as e:
        print(f"Failed to scrape {url}: {e}")
```

```python
practice_df = pd.DataFrame(all_practice_data, columns=columns_practice)
```

```python
practice_df.rename(columns={
    "Grand Prix": "Grand_Prix",
    "Pos": "Position",
    "No": "Car_Number",
    "Time": "Lap_Time",
    "Gap": "Time_Gap",
    "Laps": "Laps_Completed"
}, inplace=True)
```

```python
practice_df['Car'] = practice_df['Car'].str.strip()
```

```python
practice_df['Position'] = pd.to_numeric(practice_df['Position'], errors='coerce')
practice_df['Laps_Completed'] = pd.to_numeric(practice_df['Laps_Completed'], errors='coerce')
```

```python
practice_df = practice_df[practice_df['Position'].between(1, 20, inclusive='both')]
practice_df = practice_df[practice_df['Laps_Completed'] > 0]
```

```python
practice_df['Lap_Time'] = practice_df['Lap_Time'].fillna('N/A')
practice_df['Time_Gap'] = practice_df['Time_Gap'].fillna('N/A')
```

```python
columns_qualifying = [
    "Grand Prix","Pos","No","Driver","Car","Q1" ,"Q2" ,"Q3" ,"Laps"
]
```

```python
def scrape_qualifying_data(url):
    try:
        response = requests.get(url)
        response.raise_for_status()
        soup = BeautifulSoup(response.content, 'html.parser')

        # Extract the Grand Prix name from the URL
        grand_prix = url.split('/')[-2].replace('-', ' ').title()

        # Find the table
        table = soup.find('table', class_='f1-table f1-table-with-data w-full')
        if not table:
            print(f"Table not found for {url}")
            return []

        # Extract rows from the table
        rows = table.find('tbody').find_all('tr')
        data = []
        for row in rows:
            cells = row.find_all('td')
            if len(cells) == 8:  # Ensure the row has the correct number of columns
                data.append([
                    grand_prix,
                    cells[0].text.strip(),  # Pos
                    cells[1].text.strip(),  # No
                    cells[2].text.strip(),  # Driver
                    cells[3].text.strip(),  # Car
                    cells[4].text.strip(),  # Q1
                    cells[5].text.strip(),  # Q2
                    cells[6].text.strip(),  # Q3
                    cells[7].text.strip()   # Laps
                ])
        return data
    except requests.RequestException as e:
        print(f"Error fetching data from {url}: {e}")
        return []
```

```python
all_qualifying_data = []
```

```python
for url in urls_qualifying:
    all_qualifying_data.extend(scrape_qualifying_data(url))
```

```python
qualifying_df = pd.DataFrame(all_qualifying_data, columns=columns_qualifying)
```

```python
qualifying_df["Driver"] = qualifying_df["Driver"].str.replace(r"[A-Z]{3}$", "", regex=True).str.strip()
```

```python
qualifying_df["Car"] = qualifying_df["Car"].str.replace(r"\s+", " ", regex=True).str.strip()
```

```python
time_pattern = re.compile(r"^\d{1,2}:\d{2}\.\d{3}$")
for col in ["Q1", "Q2", "Q3"]:
    qualifying_df[col] = qualifying_df[col].apply(
        lambda x: x if pd.notnull(x) and time_pattern.match(str(x)) else None
    )
```

```python
qualifying_df["Grand Prix"] = qualifying_df["Grand Prix"].str.title().str.strip()
```

```python
qualifying_df['Pos'] = pd.to_numeric(qualifying_df['Pos'], errors='coerce')
qualifying_df["Laps"] = pd.to_numeric(qualifying_df["Laps"], errors='coerce')
```

```python
qualifying_df = qualifying_df[qualifying_df['Pos'].between(1, 20, inclusive='both')]
```

```python
def time_to_seconds(time_str):
    """Convert time in HH:MM:SS or MM:SS.SSS format to seconds."""
    parts = time_str.split(':')
    if len(parts) == 3:  # HH:MM:SS format
        hours, minutes, seconds = map(float, parts)
        return hours * 3600 + minutes * 60 + seconds
    elif len(parts) == 2:  # MM:SS.SSS format
        minutes, seconds = map(float, parts)
        return minutes * 60 + seconds
    return float(time_str)  # Just seconds

# Load the pitstop data
pitstop_df = pd.read_csv("pit_stops.csv")

# Convert the "Time" and "Total" columns to seconds
pitstop_df['Time (seconds)'] = pitstop_df['Time'].apply(time_to_seconds)
pitstop_df['Total (seconds)'] = pitstop_df['Total'].apply(time_to_seconds)

# Aggregate the data to calculate the total stops and total time in seconds for each driver in a Grand Prix
aggregated_pitstop_data = (
    pitstop_df.groupby(['Grand Prix', 'No', 'Driver', 'Car'])
    .agg(
        Total_stops=('Stops', 'sum'),
        Total_time_seconds=('Time (seconds)', 'sum')
    )
    .reset_index()
)

# Convert the total time in seconds back to HH:MM:SS or MM:SS.SSS format
def seconds_to_time(seconds):
    """Convert seconds to HH:MM:SS or MM:SS.SSS format."""
    hours = int(seconds // 3600)
    minutes = int((seconds % 3600) // 60)
    seconds = seconds % 60
    if hours > 0:
        return f"{hours:02}:{minutes:02}:{seconds:06.3f}"
    return f"{minutes:02}:{seconds:06.3f}"

aggregated_pitstop_data['Total Time'] = aggregated_pitstop_data['Total_time_seconds'].apply(seconds_to_time)

# Drop the intermediate seconds column
aggregated_pitstop_data.drop(columns=['Total_time_seconds'], inplace=True)

# Rename columns for better readability
aggregated_pitstop_data.rename(columns={
    'Total_stops': 'Total Stops'
}, inplace=True)

# Save the new DataFrame to a CSV
aggregated_pitstop_data.to_csv("aggregated_pitstops.csv", index=False, encoding="utf-8-sig")
```

```python
try:
    qualifying_df = pd.read_csv("qualifying_cleaned.csv.csv")
except FileNotFoundError:
    print("The file 'qualifying_cleaned.csv.csv' was not found.")
    raise

# Select relevant columns
columns_to_use = ['Grand Prix', 'Driver', 'Q1', 'Q2', 'Q3']
mean_df = qualifying_df[columns_to_use].copy()

# Function to convert time string to seconds
def time_to_seconds(time_str):
    if pd.isnull(time_str) or not isinstance(time_str, str):
        return np.nan
    try:
        minutes, seconds = time_str.split(':')
        seconds, milliseconds = seconds.split('.')
        total_seconds = int(minutes) * 60 + int(seconds) + int(milliseconds) / 1000
        return total_seconds
    except ValueError:
        return np.nan

# Convert Q1, Q2, Q3 to seconds
for col in ['Q1', 'Q2', 'Q3']:
    mean_df[col] = mean_df[col].apply(time_to_seconds)

# Calculate mean time in seconds, ignoring null values
mean_df['Mean Time'] = mean_df[['Q1', 'Q2', 'Q3']].mean(axis=1, skipna=True)

# Drop rows where all Q1, Q2, and Q3 are null
mean_df = mean_df.dropna(subset=['Mean Time']).reset_index(drop=True)

# Convert 'Mean Time' back to minute:second.millisecond format
def seconds_to_time(total_seconds):
    if pd.isnull(total_seconds):
        return None
    minutes = int(total_seconds // 60)
    seconds = int(total_seconds % 60)
    milliseconds = int((total_seconds % 1) * 1000)
    return f"{minutes}:{seconds:02d}.{milliseconds:03d}"

mean_df['Mean Time'] = mean_df['Mean Time'].apply(seconds_to_time)

# Save to CSV
mean_df.to_csv("qualifying_mean_times.csv", index=False, encoding="utf-8-sig")
```

```python
data = pd.read_csv('ML file(in).csv')

# Step 1: Clean column names
data.columns = data.columns.str.strip()

# Step 2: Rename columns for consistency
data.rename(columns={
    'Practice AVG Lap Time': 'Practice_Avg_Lap_Time',
    'Best Qualifycation Time': 'Best_Qualification_Time',
    'Fastest Lap': 'Fastest_Lap',
    'Fastest Lap In Seconds': 'Fastest_Lap_In_Seconds',
    'Car Number': 'Car_Number',
    'Race Laps': 'Race_Laps'
}, inplace=True)

# Step 3: Handle missing values
# Convert lap time columns to numeric format (seconds)
def convert_to_seconds(time_str):
    if pd.isna(time_str):
        return np.nan
    try:
        minutes, seconds = map(float, time_str.split(':'))
        return minutes * 60 + seconds
    except ValueError:
        return np.nan

for col in ['Practice_Avg_Lap_Time', 'Best_Qualification_Time', 'Fastest_Lap']:
    data[col + '_Seconds'] = data[col].apply(convert_to_seconds)

# Impute missing values with mean for numerical columns
data = data.assign(
    Practice_Avg_Lap_Time_Seconds=data['Practice_Avg_Lap_Time_Seconds'].fillna(data['Practice_Avg_Lap_Time_Seconds'].mean()),
    Best_Qualification_Time_Seconds=data['Best_Qualification_Time_Seconds'].fillna(data['Best_Qualification_Time_Seconds'].mean()),
    Fastest_Lap_In_Seconds=data['Fastest_Lap_In_Seconds'].fillna(data['Fastest_Lap_In_Seconds'].mean())
)

# Drop original string time columns (optional, keep if needed)
data.drop(columns=['Practice_Avg_Lap_Time', 'Best_Qualification_Time', 'Fastest_Lap'], inplace=True)

# Step 4: Check for duplicates and remove if necessary
data.drop_duplicates(inplace=True)

# Step 5: Verify data consistency
# Ensure all times are within reasonable bounds (e.g., 1 min to 2 min for lap times)
reasonable_time_bounds = (60, 120)
for col in ['Practice_Avg_Lap_Time_Seconds', 'Best_Qualification_Time_Seconds', 'Fastest_Lap_In_Seconds']:
    data = data[(data[col] >= reasonable_time_bounds[0]) & (data[col] <= reasonable_time_bounds[1])]

# Step 6: Save cleaned dataset
data.to_csv('ML_file_cleaned.csv', index=False, encoding="utf-8-sig")
```
